```python
from sklearn.model_selection import KFold
import numpy as np

X = np.arange(1, 21).reshape(10, 2)
y = np.arange(1, 11)

kf = KFold(n_splits=5)

soufian = 1

for train_idx, test_idx in kf.split(X):
    print("Fold:", soufian)
    print("Train:", train_idx, "Test:", test_idx)
    print()
    soufian += 1
```
