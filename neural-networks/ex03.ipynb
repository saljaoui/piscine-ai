```python
import numpy as np

def log_loss_custom(y_true, y_pred):
    n = len(y_true)
    loss = 0.0
    
    for yt, yp in zip(y_true, y_pred):
        loss += yt * np.log(yp) + (1 - yt) * np.log(1 - yp)
    
    return -loss / n
```

```python
import numpy as np
from sklearn.metrics import log_loss

y_true = np.array([0, 1, 1, 0, 1])
y_pred = np.array([0.1, 0.8, 0.6, 0.5, 0.3])

custom_loss = log_loss_custom(y_true, y_pred)
sklearn_loss = log_loss(y_true, y_pred)

print("Custom", custom_loss)
print("Sklearn", sklearn_loss)
```
