```python
from neural_network import Neuron
import numpy as np

neuron_h1 = Neuron(0.05, 0.001, 0)
neuron_h2 = Neuron(0.02, 0.003, 0)
neuron_o1 = Neuron(2,0,0)

data = [
    (12, 15, 1, "Bob"),
    (10, 9, 0, "Eli"),
    (18, 18, 1, "Tom"),
    (13, 14, 1, "Ryan")
]

y_true = []
y_pred = []

for math_g, chem_g, y, name in data:
    h1_out = neuron_h1.feedforward(math_g, chem_g)
    h2_out = neuron_h2.feedforward(math_g, chem_g)
    y_hat = neuron_o1.feedforward(h1_out, h2_out)

    y_true.append(y)
    y_pred.append(y_hat)
    print(name + ":" , y_hat)
```

```python
loss = -np.mean(
    np.array(y_true) * np.log(y_pred) +
    (1 - np.array(y_true)) * np.log(1 - np.array(y_pred))
)
loss
```
